```python
import numpy as np 
import pandas as pd
from scipy.stats import stats
import math
```

```python
cutlet = pd.read_csv('C:\\Nikhil\\\DS\\ML\\Assignments\\Hypothesis testing\\Cutlets.csv')
cutlet.head()
```

```python
cutlet.isnull()
```

```python
# There is no null values so we can apply hypothesis testing
# Now here, our null hypothesis is "there is no difference in the diameter of both the samples"
# And alternate hypothesis is "there is some difference between the diameter of both the samples"
```

```python
sample_1 = cutlet['Unit A']
sample_2 = cutlet['Unit B']
```

```python
ttest,p_value = stats.ttest_ind(a=sample_1, b=sample_2)
```

```python
print(p_value)
```

```python
# Hence
if p_value > 0.05 :
    print ('We accept null hypothesis and rejects alternate hypothesis')
else:
    print ('We rejects null hypothesis and accept alternate hypothesis')
```
