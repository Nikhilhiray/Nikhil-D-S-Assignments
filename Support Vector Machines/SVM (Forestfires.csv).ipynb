```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
df = pd.read_csv('C:\\Nikhil\\DS\\Assignments\\Support Vector Machine\\forestfires.csv')
df.head()
```

```python
df.info()
```

```python
df['size_category'].value_counts()
```

```python
# Changing large to '1' and small to '0' 
df['size_category'] = np.where(df['size_category'].str.contains("large"), 1, 0)
df.head()
```

```python
# We have to drop month and day column because we already have dummy variables
df.drop(['month','day'], axis=1, inplace = True)
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:,:28]
Y = df.iloc[:,28]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

# Randomized search CV

```python
clf = SVC()
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001] }]
gsv = RandomizedSearchCV(clf,param_grid,cv=10,verbose=300)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

# Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001] }]
gsv = GridSearchCV(clf,param_grid,cv=10,verbose=300)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

# Applying SVM classification

```python
# In both randomized search CV and grid search CV the 
# parameter are same ('C': 10, 'gamma': 0.5, 'kernel': 'linear') So,
clf = SVC(kernel = 'linear', C = 10, gamma = 0.5)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
