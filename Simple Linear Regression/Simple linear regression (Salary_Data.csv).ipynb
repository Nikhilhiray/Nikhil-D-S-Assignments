```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model 
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
#importing the data 
salary = pd.read_csv("C:\\Nikhil\\ML\\Assignments\\Simple linear regresion\\Salary_Data.csv")
salary
```

```python
# implementing 'OLS=Ordinary Least Squares' function to build model 
model = smf.ols("Salary~YearsExperience", data = salary).fit()
```

```python
# To see the best fit line
sns.regplot(x="YearsExperience", y="Salary", data = salary)
```

```python
# Cofficients (intercept & sloap)
model.params
```

```python
# All predicted values of salary
model.fittedvalues
```

```python
# Adding predicted values to the dataframe 
salary['Predicted Salary'] = model.fittedvalues
salary
```

```python
# Checking the difference between actual salary and predicted salary (errors)  
salary["Error"] = model.resid
salary
```

```python
model.rsquared
#95.69% of variation in the dependent variable is explained by the given independent variable
```

```python
model.pvalues
```
